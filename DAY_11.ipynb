# **Evaluation Matrics**

```python
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

#Actual value prediction
y_true = np.array([50,60,75,90,100])

#predicted value
y_pred = np.array([52,58,78,88,105])

#MAE(mean_absoute_error)
mae= mean_absolute_error(y_true,y_pred)
print("MAE:",mae)

#MSE(mean_square_error)
mse = mean_squared_error(y_true,y_pred)
print("MSE:",mse)

#RMSE(root_mean_square_error)
rmse = np.sqrt(mse)
print("RMSE:",rmse)

#r2 score
r2 = r2_score(y_true,y_pred)
print("R2 Score:",r2)
```

```python
import matplotlib.pyplot as plt

# graph of the mae, mse, rmse,r2
plt.y_true = np.array([50,60,75,90,100])
plt.y_pred = np.array([52,58,78,88,105])

plt.plot(y_true,y_pred)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Evaluation Matrics")
plt.grid(True)

# # representing  mae, mse, rmse,r2 on graph
# plt.title(mae,mse,rmse,r2)
plt.legend(["MAE","MSE","RMSE","R2"])

plt.show()
#
```

### Lasso Regression (L1 Regularization)
Lasso (Least Absolute Shrinkage and Selection Operator) regression is a type of linear regression that uses shrinkage. Shrinkage methods involve penalizing the sum of absolute values of the coefficients (L1 penalty) to reduce their magnitude and perform feature selection. It can force some coefficients to be exactly zero, effectively removing the corresponding features from the model. This makes Lasso useful for models with a large number of features or when feature selection is desired.

### Ridge Regression (L2 Regularization)
Ridge regression is another regularization technique for linear regression. It adds a penalty equivalent to the square of the magnitude of the coefficients (L2 penalty). This penalty term helps to reduce the complexity of the model by shrinking the coefficients towards zero, but it rarely forces them to be exactly zero. Ridge regression is effective in preventing overfitting, especially when there is multicollinearity (high correlation between independent variables).

```python
#Lasso R and Ridge R
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_absolute_error

data = {
    "size":[1000,1200,1500,1800,2000,2200,2500,2700,3000,3200],
    "bedrooms":[2,2,3,3,3,4,4,4,5,5],
    "age":[20,18,15,12,10,8,6,5,4,2],
    "price":[30,35,45,50,55,60,68,72,80,85]
}
df = pd.DataFrame(data)
print(df)

# data split in x and y.
x = df[['size','bedrooms','age']]
y = df['price']

#split data in training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#training -> ridge reggression
ridge = Ridge(alpha=1.0)
ridge.fit(x_train,y_train)

#prediction
ridge_p = ridge.predict(x_test)
print(ridge_p)
print(x_test)
print("prediction")
print(ridge_p)
print("actual")
print(y_test)

#evaluation
mae = mean_absolute_error(y_test,ridge_p)
print("MAE:",mae)

mse = mean_squared_error(y_test,ridge_p)
print("MSE:",mse)

#model intercept
print(ridge.intercept_)

# model intercept
print(ridge.intercept_)

# model coef price = intercept + (cof1,* feat1)
print(ridge.coef_)
```
