```python
import pandas as pd

data = {
    "Emp_ID" : [101,102,103,104,105,106],
    "Name" :["Amit",'Riya','Raj','Sara','John','Neha'],
    'Department' : ['HR','IT','Finance','IT','HR','Finance'],
    'Salary' : [60000,70000,50000,80000,55000,65000],
    "Experience" : [2,3,5,4,1,3]
}
df = pd.DataFrame(data)
print(df,"\n")
print('------All PANDAS Functions------')
print('Info :')
print(df.info,"\n")
print('Head :')
print(df.head(),"\n")
print('Tail :')
print(df.tail(),"\n")
print('Shape :')
print(df.shape,"\n")
print('Rename :')
print(df.rename(columns={"Name":"Emp_name"},inplace=True))
print('\n')
print('Describe :')
print(df.describe())
print('\n')
print('Describe Numeric :')
print(df.describe(include='number'))
print('\n')
print('Describe All :')
print(df.describe(include='all'))
```

# 4. **Numerical Operations & Transformations**
*(e.g., salary increment calculations)*

```python
import pandas as pd # Import pandas library

d = { # Create dictionary for DataFrame
"name":["vishal","virat","vineet"],
"salary":[100,200,300]
}
df = pd.DataFrame(data=d) # Create DataFrame
df["marks"] = df["salary"] / 2 # Calculate marks
df # Display initial DataFrame

# Increment salary by 10%
print('Original DataFrame')
print(df,'\n\n')
print('DataFrame with incremented salaries')
df['salary'] = df['salary'] * 1.10 # Apply 10% salary increment
df # Display updated DataFrame
```
