# Types of Operators_2

```python
#Example of Arithmetic operator(%)
num_1 = int(input('Enter the number :'))
num_2 = 2
print(num_1 % num_2)
```

```python
#2. comparison(Relational) Operators --> Answer comes in True or False or (Boolean values).
a = 35
b = 55
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
print(a > b)
print(a < b)
```

```python
#3. Assignment Operators
Alpha = 58
Alpha = Alpha + 8
print('Alpha 1 : ', Alpha)
Alpha = Alpha - 8
print('Alpha 2 : ',Alpha)
Alpha += 8
print('Alpha 3 : ',Alpha)
Alpha -= 8
print('Alpha 4 : ',Alpha)
Alpha *= 8
print('Alpha 5 : ',Alpha)
Alpha /= 8
print('Alpha 6 : ',Alpha)
Alpha %= 8
print('Alpha 7 : ',Alpha)
Alpha //= 8
print('Alpha 8 : ',Alpha)
Alpha **= 8
print('Alpha 9 : ',Alpha)
```

```python
# 4. Logical Operator --not
a = 20
b = 40
print(not False)
print(not True)
print(not (a>b) )
print(not (b>a) )
```

```python
#--and--
num1 = True
num2 = False
print('After performing AND operator :', num1 and num2)
num2 = True
print('After performing AND operator :', num1 and num2,'\n')

#Example using integer values
a = 10
b = -10
print(a and b)
b = 0
print(a and b,'\n')

#2. Example (and, or, and not)
name1 = 'hello'         # In string 'name' --> True
name2 = ''              # in blank string('') --> False
print('hello and blank :',name1 and name2)
print('hello or blank :',name1 or name2)
print('hello and not blank :',name1 and not name2)
```

```python
print('After performing OR operator :', num1 or num2)
print('OR Operator :',(a==b)or(b>a))    # OR operator will be true when one of the condition is true.

#Example
a = 10
b = -10
print(a or b)
b = 0
print(a or b ,'\n')

# Example for string concept
name1 = 'hello'
name2 = ''
print('hello or blank :',name1 or not name2)
print('not hello or blank :', not name1 or name2)

```

**5. Bitwise Operator**

```python
# &
a = 5 # 101
b = 2 # 010
print("5 & 2 :",a & b)

#Examples of bitwise operators
a = 10
b = 20
print('a >> b :',a >> b)
print("a << b :",a << b)
print("a & b :",a & b)
print("a | b :",a | b)
print("a ^ b :",a ^ b)
```

**6. Identity Operators**


```python
# is | is not

# example 1
name = 10
name1 = 10
print(name == name1 )# value

print(name is name1) # value | address
print(id(name))
print(id(name1))

#example   (-5 to 256) -> both variable point on same address
num1 = [1,2,3]
num2 = [1,2,3]
num3 = num1

print(num1 == num2)

print(num1 is num2)
print(id(num1))
print(id(num2))
print(id(num3))
```
