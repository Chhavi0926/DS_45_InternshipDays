```python
#2. comparison(Relational) Operators
a = 35
b = 55
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
print(a > b)
print(a < b)
```

```python
#2. Assignment Operators
Alpha = 58
Alpha = Alpha + 8
print('Alpha 1 : ', Alpha)
Alpha = Alpha - 8
print('Alpha 2 : ',Alpha)
Alpha += 8
print('Alpha 3 : ',Alpha)
Alpha -= 8
print('Alpha 4 : ',Alpha)
Alpha *= 8
print('Alpha 5 : ',Alpha)
Alpha /= 8
print('Alpha 6 : ',Alpha)
Alpha %= 8
print('Alpha 7 : ',Alpha)
Alpha //= 8
print('Alpha 8 : ',Alpha)
Alpha **= 8
print('Alpha 9 : ',Alpha)
```

```python
# 4. Logical Operator --not
a = 20
b = 40
print(not False)
print(not True)
print(not (a>b) )
print(not (b>a) )
```

```python
#--and--
num1 = True
num2 = False
print('After performing AND operator :', num1 and num2)
num2 = True
print('After performing AND operator :', num1 and num2)
```

```python
print('After performing OR operator :', num1 or num2)
print('OR Operator :',(a==b)or(b>a))    # OR operator will be true when one of the condition is true.
```
